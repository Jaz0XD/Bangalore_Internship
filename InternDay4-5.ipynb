```python
gender_map={'Male':1, 'Female': 0}
Education_map = {'Graduate': 1, 'Not Graduate': 0}
Self_Employed_map = {'Yes':1, 'No': 0}
Married_map = {'Yes':1, 'No': 0}
Property_Area_map = {'Urban': 2, 'Semiurban': 1, 'Rural':0}
dependes_map = {'1': 1,'2':2, '3+':3, '0': 0}
```

```python
df['Gender'] = df['Gender'].map(gender_map)
df['Education'] = df['Education'].map(Education_map)
df['Self_Employed'] = df['Self_Employed'].map(Self_Employed_map)
df['Married'] = df['Married'].map(Married_map)
df['Property_Area'] = df['Property_Area'].map(Property_Area_map)
df['Dependents'] = df['Dependents'].map(dependes_map)
```

```python
df.head()
```

```python
x = df.drop(columns=['Loan_ID', 'Loan_Status'])
```

```python
x
```

```python
y=df['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

# Day 5
### Model Building 

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
```

```python
df.shape
```

```python
ranf_forest = RandomForestClassifier(n_estimators=100, max_depth = 3)
```

```python
ranf_forest.fit(x_train, y_train)
```

```python
y_pred = ranf_forest.predict(x_test)
```

```python
y_pred
```

```python
accuracy_score(y_pred, y_test)
```

```python
confusion_matrix(y_pred, y_test)
```

```python
print(classification_report(y_pred, y_test))
```

#### Precision = TP / TP + FP

#### Recall = TP / TP + FN

###### A good model should have higher precision and recall
