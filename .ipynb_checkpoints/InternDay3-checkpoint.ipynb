## What are the top 3 bedroom counts with the highest average house prices? Plot them with horizontal bars

### Method 1

```python
df.head()
```

```python
plt.figure(figsize=(8,5))
```

```python
top_bedrooms = df['bedrooms'].value_counts().nlargest(3).index
filtered_df = df[df['bedrooms'].isin(top_bedrooms)]
```

```python
sns.barplot(x='price', y='bedrooms', data=filtered_df, estimator=np.mean, orient='h',)
plt.title('Top 3 Average Price by Number of bedrooms')
plt.show()
```

### Method 2

```python
temp = df.groupby('bedrooms')['price'].mean().reset_index()
temp.head()
```

```python
temp.sort_values(by='price', ascending=False)
```

```python
top3 = temp.sort_values(by='price', ascending=False).head(3)
```

```python
top3
```

```python
sns.barplot(x='price', y='bedrooms', data=top3, estimator=np.mean, orient='h')
plt.title('Top 3 Average Price by Number of bedrooms')
plt.show()
```

## Does the presence of basement significantly shift the distribution of price or not?

```python
sns.histplot(x='price',hue='basement', data=df, bins=20, kde=True)
plt.title("Price shift distribution based on basement")
plt.xlabel('Prices')
plt.ylabel('Basement')
plt.show() #To remove the text
```

```python
plt.figure(figsize=(6,6))
df['furnishingstatus'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Furnishing Status Distribution')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.kdeplot(data=df, x='price', hue='furnishingstatus' ,fill=True)
plt.title('KDE Plot of House Prices')
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(14,5))
sns.kdeplot(data=df, x='price', hue='furnishingstatus', fill = True, ax = axes[1])
sns.boxplot(data=df , x ='furnishingstatus', y = 'price', ax=axes[0])


# Setting title and label For Graph 0
axes[0].set_title('Distribution of House Prices')
axes[0].set_xlabel('Price')
axes[0].set_ylabel('Count')


# Setting title and label For Graph 1
axes[1].set_title('KDE Plot of House Prices')

plt.show()
```
